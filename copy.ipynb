```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

%matplotlib inline
```

```python
#ティッカーシンボルを引数に指定してとってくる
aapl = yf.Ticker('AAPL')
```

```python
aapl.history()
```

```python
aapl.history().shape
```

```python
#50日分のデータ取得
days = 50
hist = aapl.history(period=f'{days}d')
hist
```

```python
#Dateがカラムにないことがわかる　つまり　Dateはインデックス
hist.columns
```

```python
#hist.reset_index()
```

```python
hist_msft = yf.Ticker('MSfT').history(period=f'{days}d')
hist_msft.head()
```

```python
#histとhist_msftを合わせる
pd.concat([hist,hist_msft],axis = 1).head()
```
