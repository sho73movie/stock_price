```python
days = 20
#key:会社名　values:ティッカーシンボル
tickers = {
    'apple' : 'AAPL',
    'facebook' : 'FB'
}
company = 'facebook'

tkr = yf.Ticker(tickers[company])
hist = tkr.history(period=f'{days}d')
hist.index = hist.index.strftime('%d %B %Y')
hist = hist[['Close']]
hist.columns = [company]
hist = hist.T
hist.index.name = 'Name'
hist
```

```python
def get_data(days,tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
    # company = 'facebook'

        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df,hist])
    return df
```

```python
days = 20
tickers = {
    'apple' : 'AAPL',
    'facebook' : 'FB',
    'google':'GOOGL',
    'microsoft':'MSFT',
    'netflix':'NFLX',
    'amazon':'AMZN'
}

get_data(days,tickers)
```

```python
import altair as alt
```

```python
df
```

```python
companies = ['apple','facebook']
data = df.loc[companies]
data
```

```python
#アルファベット順
data.sort_index()
```

```python
data = data.T.reset_index()
data.head()
```

```python
#melt関数を使用
data = pd.melt(data,id_vars=['Date'])
data
```

```python
data = pd.melt(data,id_vars=['Date']).rename(
    columns = {'value':'Stock Prices(USD)'}
)
data
```

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8)
    .encode(
        x='Date:T',
        y=alt.Y('Stock Prices(USD):Q', stack = None),
        color='Name:N'
    )
)
```

```python
chart
```

```python
#スケールの変更 y軸　
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8)
    .encode(
        x='Date:T',
        y=alt.Y('Stock Prices(USD):Q', stack = None, scale = alt.Scale(domain=[200,350])),
        color='Name:N'
    )
)
```

```python
chart
```

```python
#グラフの外にはみ出る現象を修正 clip=True
ymin,ymax = 150,340
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8,clip=True)
    .encode(
        x='Date:T',
        y=alt.Y('Stock Prices(USD):Q', stack = None, scale = alt.Scale(domain=[ymin,ymax])),
        color='Name:N'
    )
)
```
